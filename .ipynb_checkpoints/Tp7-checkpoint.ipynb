```python
# -*- coding: utf-8 -*-
"""
Éditeur de Spyder

Ceci est un script fait pour un TP de polytech adapté depuis.
https://towardsdatascience.com/keras-transfer-learning-for-beginners-6c9b8b7143e

Auteur : Romain Raveaux & Yongda LIN & Qingling MENG
"""

from keras.layers import Dense,GlobalAveragePooling2D
from keras.applications import MobileNet
from keras.applications.mobilenet import preprocess_input
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
import tensorflow as tf

import os

#Some parameters that depends on the final application
#number of classes 
#the number of iterations
imagedir = "Images"
nbclasses = len(os.listdir(imagedir))
nbepochs = 10
nblayer = 10

#We create the mode from an already trained model called MobileNet
#We do not take the last layer of the model that is dedicated to recognize 1000 classes 
base_model=MobileNet(input_shape=[224,224,3],weights='imagenet',include_top=False) 
#imports the mobilenet model and discards the last 1000 neuron layer.
x=base_model.output

#We modify the last layer to adapt it to our problem
x=GlobalAveragePooling2D()(x)
#x=Dense(1024,activation='relu')(x) #we add dense layers so that the model can 
#learn more complex functions and classify for better results.
preds=Dense(nbclasses,activation='softmax')(x) 
#final layer with softmax activation



#specify the inputs
#specify the outputs
#now a model has been created based on our architecture
model=Model(inputs=base_model.input,outputs=preds)



#some printing of the names of the layers
for i,layer in enumerate(model.layers):
    print(i,layer.name)
    
# ---------------    5 last

#We do not have enough data to retrain from scrath a big neural netqork  with 88 layers.
#We will retrain only the last layer, the last past
for layer in model.layers:
    layer.trainable=False
# or if we want to set the first 87 layers of the network to be non-trainable
for layer in model.layers[:(88 - nblayer)]:
    layer.trainable=False
for layer in model.layers[(88 - nblayer):]:
    layer.trainable=True
    
```

Cross-validation

```python
from keras.preprocessing.image import ImageDataGenerator

batch_size = 32
#We set the optimization problem and the way to solve it
# Adam optimizer
# loss function will be categorical cross entropy
# evaluation metric will be accuracy
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])

# Leave-one-out cross-validation (5-fold)
print("Layer\t", str(nblayer), "Class\t", str(nbclasses))

data_datagen = ImageDataGenerator(
    rescale = 1. / 255,
    shear_range = 0.2,
    zoom_range = 0.2,
    validation_split = 0.2,
    horizontal_flip = True)

training_set = data_datagen.flow_from_directory(
               imagedir,
               target_size = (224,224),
               batch_size = batch_size,
               class_mode = 'categorical',
               subset = "training")

test_set = data_datagen.flow_from_directory(
               imagedir,
               target_size = (224,224),
               batch_size = batch_size,
               class_mode = 'categorical',
               subset = "validation") 

model.fit_generator( 
    training_set, 
    steps_per_epoch=batch_size, 
    epochs=nbepochs, 
    validation_data=test_set)

# model.fit_generator(generator=train_generator,
#                    steps_per_epoch=step_size_train,epochs=nbepochs)
```

Save model

```python

#Save the model : weights and architectures of the neural network
model.save('modelpolytech.h5')

#Convert the model from keras file to tflite file dor android studio
converter = tf.lite.TFLiteConverter.from_keras_model_file( 'modelpolytech.h5')
tflitemodel = converter.convert()
open ("modelpolytech.tflite" , "wb") .write(tflitemodel)
```
